Exploración básica con una API de laboratorio.

```python
import requests
import pandas as pd
import json
```

```python
# 1. Consulta la API pública JSONPlaceholder utilizando el método GET para obtener:
# Lista de publicaciones (/posts)
# Lista de usuarios (/users)
# Lista de tareas (/todos)

# URLs de la API
url_posts = "https://jsonplaceholder.typicode.com/posts"
url_users = "https://jsonplaceholder.typicode.com/users"
url_todos = "https://jsonplaceholder.typicode.com/todos"

# Peticiones GET
res_posts = requests.get(url_posts)
res_users = requests.get(url_users)
res_todos = requests.get(url_todos)

# Convertir a JSON
posts = res_posts.json()
users = res_users.json()
todos = res_todos.json()

# 2. Muestra por pantalla:

# La cantidad total de cada recurso.
print("Cantidad total de posts:", len(posts))
print("Cantidad total de usuarios:", len(users))
print("Cantidad total de tareas:", len(todos))

# El código de estado de cada petición.
print("Código de estado /posts:", res_posts.status_code)
print("Código de estado /users:", res_users.status_code)
print("Código de estado /todos:", res_todos.status_code)
```

```python
# 3. Realiza una petición a una publicación inexistente para obtener un error 404 y muestra el código de estado recibido.
peticion = requests.get("https://jsonplaceholder.typicode.com/posts/999999")

print("codigo de estado:", peticion.status_code)
```

```python
# 4. Realiza una petición POST para crear una nueva publicación ficticia. Incluye el título, el cuerpo del mensaje y un userId.
# Muestra la respuesta JSON.
# Muestra el código de estado.

nueva_publicacion = {
    "title": "Mi publicación ficticia",
    "body": "Este es el contenido del post",
    "userId": 1
}

respuesta_post = requests.post(
    "https://jsonplaceholder.typicode.com/posts",
    json=nueva_publicacion
)

print("Respuesta JSON:", respuesta_post.json())
print("Código de estado:", respuesta_post.status_code)
```

```python
# 5. Realiza una petición PATCH para modificar parcialmente una publicación existente.
# Muestra la respuesta JSON.
# Muestra el código de estado.

modificacion = {
    "title": "Título actualizado"
}

respuesta_patch = requests.patch(
    "https://jsonplaceholder.typicode.com/posts/1",
    json=modificacion
)

print("Respuesta JSON:", respuesta_patch.json())
print("Código de estado:", respuesta_patch.status_code)
```

```python
# 6. Haz una petición DELETE sobre una publicación.
# Muestra la respuesta JSON.
# Muestra el código de estado.

respuesta_delete = requests.delete("https://jsonplaceholder.typicode.com/posts/1")

print("Respuesta JSON:", respuesta_delete.json())
print("Código de estado:", respuesta_delete.status_code)
```

# Nivel 3
API de Open Data Barcelona

```python
# 1. Utiliza la API de Open Data BCN para:
# Buscar un dataset de tu interés mediante package_search
url_search = "https://opendata-ajuntament.barcelona.cat/data/api/3/action/package_search"

param_busqueda = {"q": "transport"}

respuesta = requests.get(url_search, params=param_busqueda).json()

# Mostrar el primer dataset encontrado
primer_dataset = respuesta["result"]["results"][0]
print("Primer dataset encontrado:")
print(json.dumps(primer_dataset, indent=4))


# package_show
dataset_id = primer_dataset["id"]

url_show = "https://opendata-ajuntament.barcelona.cat/data/api/3/action/package_show"
respuesta_show = requests.get(url_show, params={"id": dataset_id}).json()

# print(respuesta)
print(respuesta_show)
```

```python
# 2. De los resultados obtenidos, selecciona uno que tenga recursos disponibles en CSV o JSON
recurso_id = respuesta_show["result"]["resources"][0]["id"] 

url_data = "https://opendata-ajuntament.barcelona.cat/data/api/3/action/datastore_search"


# 3. Con el recurso_id del recurso seleccionado, realiza una consulta mediante datastore_search por:
# Recuperar al menos 100 registros.
param_datastore = {
    "resource_id": recurso_id,
    "limit": 100
}

data_respuesta = requests.get(url_data, params=param_datastore).json()

records = data_respuesta["result"]["records"]
print(len(records))
print(json.dumps(records[:100], indent=4))
```

```python
# 4. Convierte los resultados a un DataFrame.
df = pd.DataFrame(records)
df.head()
```

```python
# 5. Guarda DataFrame en un archivo .csv.
df.to_csv("barcelona_dataset.csv", index=False)
```
