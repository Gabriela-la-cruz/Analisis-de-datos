## 1. Calculadora del √≠ndice de masa corporal
Escribe una funci√≥n que calcule el IMC ingresado por el usuario/a, es decir, quien lo ejecute deber√° ingresar estos datos. Puedes obtener m√°s informaci√≥n de su c√°lculo en: üîó √çndice de masa corporal IMC que es y c√≥mo se calcula.
La funci√≥n debe clasificar el resultado en sus respectivas categor√≠as.

Consejo: Intenta validar los datos previamente, para que env√≠e un mensaje de advertencia si los datos introducidos por el usuario/a no est√°n en el formato adecuado o no toma valores razonables.

```python
# Pedir datos al usuario 
peso = input("Dime tu peso en kg")
altura = input("Dime cuanto mides en metros")

def calculadora (peso,altura):   
    # Intentar convertir los valores a n√∫meros 
    try:
        peso = float(peso)
        altura = float(altura)
    except:
        print("Formato no valido")
        # Validar rango de peso  y altura
    else:
        if peso <= 1 or peso > 250:
            print("No es un peso valido")
            return
    if altura < 0.40 or altura >2.50:
        print("Altura no valida")
        return
    # Calcular IMC 
    imc = round(peso /(altura * altura),2) 
    # Clasificar el IMC
    if imc < 18.5:
        return "Estas bajo de peso"
    elif 18.5 <=  imc <= 24.9:
        return "Tienes un peso normal"
    elif 25 <= imc <= 29.9: 
        return "Tienes sobrepeso"
    elif imc > 30:
         return "Tienes un problema de obesidad"
# Ejecutar la funcion y mostrar el resultado 
masa_corporal= calculadora(peso,altura)
print(masa_corporal)
```

## 2. Conversor de temperaturas
Existen diversas unidades de temperatura utilizadas en distintos contextos y regiones. Las m√°s comunes son Celsius (¬∞C), Fahrenheit (¬∞F) y Kelvin (K). Tambi√©n existen otras unidades como Rankine (¬∞Ra) y R√©aumur (¬∞Re).
Selecciona al menos 2 conversores, de modo que al introducir una temperatura devuelva, como m√≠nimo, dos conversiones, de modo que se puedan guardar (recuerda que un print() no se puede guardar nunca).

Consejo: Intenta validar los datos previamente, para que env√≠e un mensaje de advertencia si los datos introducidos por el usuario/a no est√°n en el formato adecuado.
(EXTRA): Piensa una manera de almacenar todas las posibles conversiones en un solo objeto (¬øLista? ¬øDiccionario? ¬øDataFrame?) en lugar de escribir muchos if else en funci√≥n de la temperatura de origen y la temperatura de destino.

```python
temperatura = input("Dime la temperatura")
unidad = input("En que unidad de temperauta esta? ")

def conversor(temp,unid):
    resultados = {}
    unidades = "F","K","C"
    # Diccionario para que eval pueda usar la temperatura como F, C o K
    u = {"F": temp, "C": temp, "K": temp}
    # Validar formato
    try:
        temp = float(temp)
    except:
        print("Error: la temperatura debe ser un n√∫mero")
        return None
    # Convertir el texto a mayuscula
    unid = unid.upper()
    # Validar unidad
    if unid not in unidades:
        print("Error: unidad no valida, debe ser C, K o F.")
        return None
    # Diccionario de conversiones
    conversiones = {"F":{"C": "(F - 32) * 5/9", "K": "(F - 32) * 5/9 + 273.15"},
                   "C":{"F": "(C * 9/5) + 32", "K": "C + 273.15"},
                   "K":{"C": "K - 273.15","F": "(K - 273.15) * 9/5 + 32"}
    }
    # Aplicar cada f√≥rmula y guardar el resultado
    for clave, formula in conversiones[unid].items():
        resultados[clave] = round(eval(formula, {}, u), 2)

    return resultados 

resultado = conversor(temperatura, unidad)
if resultado is None: 
    print("No se pudo realizar la conversi√≥n.")
else: 
    print(resultado)
```

## 3. Palabra contador de un texto.

Escriba una funci√≥n que d√© un texto, muestre las veces que aparece cada palabra. Trate de manejar todos los casos posibles que hacen que el programa no funcione correctamente.

(EXTRA): ¬øCu√°l es la longitud promedio de las palabras en el texto que escribi√≥? ‚ÄúHola, ¬øc√≥mo va todo?‚Äù Debe regresar (4+3+2) / 3 = 3

```python
def contar_palabras(palabras):
    # Crear diccionario con la frecuencia de cada palabra
    diccionario = {}
    for elemento in palabras:
        if elemento in diccionario:
            diccionario[elemento] += 1
        else:
            diccionario[elemento] = 1
    return diccionario
# Pedir frase y eliminar espacios al inicio y al final
try: 
    frase = input("Dime una frase: ").strip()
    # Validar que la frase no este vacia
    if not frase:
        raise ValueError("La frase no puede estar vacia")

    # Quitar signos de puntuaci√≥n
    signos = ",.;:¬ø?¬°!\"'"
    for s in signos:
        frase = frase.replace(s, "")

    # Convertir a minuscula la frase y separar en palabras 
    palabras = frase.lower().split()

    # Verificar que haya palabras despu√©s de limpiar el texto
    if not palabras:
        raise ValueError("No hay palabras v√°lidas")

    # Calcular la longitud de cada palabra
    longitudes = [len(p) for p in palabras]

    # Calcular el promedio de la longitud
    promedio = round(sum(longitudes) / len(longitudes), 2)
    
    respuesta = contar_palabras(palabras)

    print("Promedio de longitud:", promedio)
    print("Conteo de palabras:", respuesta)

except ValueError as e:
    # Mostrar mensaje de error si ocurre una validacion fallida
    print("Error:", e)
```

# 4. Diccionario inverso (con posibilidad de duplicados)

Resulta que el cliente tiene una encuesta muy antigua que se almacena en un diccionario y los resultados los necesitan a la inversa, es decir, el intercambio de las claves y valores. Los valores y las claves en el diccionario original son √∫nicos; si este no es el caso, la funci√≥n debe imprimir un mensaje de advertencia, junto con una lista con los valores asociados con la clave repetida.

```python
libros = {
    "El Quijote": "1605",
    "La metamorfosis": "1915",
    "1984": "1949",
    "Ulises": "1922",
    "Orgullo y prejuicio": "1915"
}

def reverse_dictionary(d):
    dictionary = {}
    for key, value in d.items():
        # Avisar de valor repetido
        if value in dictionary:
            print(f"El valor '{value}' esta repetido en las claves:{dictionary[value] + [key]}")
            # Diccionario invertido
        dictionary[value] = [key]
    return dictionary

reverse_dictionary(libros)
```

# 1. Contador de palabras y ordenador de palabras de un texto.

El cliente estaba contento con el contador de palabras, pero ahora quiere leer archivos TXT y calcular la frecuencia de cada palabra ordenada dentro de las entradas habituales del diccionario seg√∫n la letra con la que comienzan, es decir, las claves deben ir de A a Z y dentro de A debemos ir de A a Z. 

```python
def procesar_texto(ruta):
    with open(ruta, "r", encoding="utf-8") as f:
        texto = f.read().lower()

    # Quitar signos de puntuaci√≥n
    for s in ",.;:¬ø?¬°!\"'()":
        texto = texto.replace(s, "")

    palabras = texto.split()

    # Contar palabras sin usar get
    frecuencia = {}
    for palabra in palabras:
        if palabra in frecuencia:
            frecuencia[palabra] += 1
        else:
            frecuencia[palabra] = 1

    # Agrupar por letra inicial
    agrupado = {}
    for palabra in sorted(frecuencia):
        inicial = palabra[0]

        if inicial not in agrupado:
            agrupado[inicial] = {}

        agrupado[inicial][palabra] = frecuencia[palabra]

    return agrupado
procesar_texto(r"C:\Users\GCCM_\OneDrive\Desktop\Bootcamp\tu me quieres blanca.txt")
```

# 2. Conversi√≥n de tipos de datos.

El cliente recibe una lista de datos y necesita generar dos listas, la primera donde estar√°n todos los elementos que pudieron convertirse en flotantes y la otra donde est√°n los elementos que no pudieron convertirse. 


```python
conversor = ['1.3', 'one' , '1e10' , 'seven', '3-1/2',('2',1,'not-a-number'), [1,2,'3','3.4']]

def conversor_numeros(lista):
    convertibles = []
    no_convertibles = []

    for elemento in lista:
        # Si el elemento es una lista o tupla, recorremos sus elementos
        if isinstance(elemento, (list, tuple)):
            for sub_el in elemento:
                try:
                    sub_el = float(sub_el)
                    convertibles.append(sub_el)
                except:
                    no_convertibles.append(sub_el)
        # Si el elemento no es lista/tupla, intentamos convertirlo directamente
        else:
            try:
                elemento = float(elemento)
                convertibles.append(elemento)
            except:
                no_convertibles.append(elemento)

    return convertibles, no_convertibles

conversor_numeros(conversor)
```

# 1. Generador de contrase√±as

Explora el funcionamiento del m√≥dulo aleatorio de la biblioteca numpy. üîó M√≥dulo aleatorio en NumPy

En este punto, el cliente ha detectado un problema con las contrase√±as utilizadas por sus trabajadores. Asdf1234, fechas de cumplea√±os o similares. Para solucionar esto, hemos sido comisionados por una funci√≥n Python que genera contrase√±as m√°s seguras. La funci√≥n debe depender de los siguientes par√°metros:

    Longitud (int): Longitud de la contrase√±a

    Mayusculas (bool = True): Si aparecen may√∫sculas

    Min√∫sculas (bool = Verdadero): Si aparecen min√∫sculas

    N√∫meros (bool = True): Si los n√∫meros van a aparecer
    
    signos (bool = Falso): Si los caracteres especiales van a aparecer (,-$? o similar)

As√≠, si ejecutamos la funci√≥n de la siguiente manera:
create_password(10, Verdadero, Verdadero, Verdadero)
Debemos obtener una salida (que debemos poder guardar) del estilo:
9Er,5Vn8P$
Aseg√∫rese de que se cumplan todos los criterios y que estas contrase√±as sean realmente aleatorias.

(EXTRA) Explorar c√≥mo podr√≠amos hacer que la funci√≥n copia autom√°ticamente la contrase√±a al soporte de papel de la computadora (como si la hubi√©ramos seleccionado y hecho ctrl + copia).


```python
from numpy import random
import pyperclip

def contrase√±a_segura(longitud,
                    mayusculas=True,
                    minusculas=True,
                    numeros=True,
                    signos=False):
 
    # Definir caracteres posibles
    caracteres = {
        "mayusculas": "ABCDEFGHIJKLMNOPQRSTUVWXYZ",
        "minusculas": "abcdefghijklmnopqrstuvwxyz",
        "numeros": "0123456789",
        "signos": "!@#$%&?-_,"
    }
    # Lista para construir la contrase√±a 
    contrase√±a = []

    # String con caracteres permitidos 
    caracteres_permitidos = ""

    # Garantizar que se use al menos un caracter permitido 
    if mayusculas:
        contrase√±a.append(random.choice(list(caracteres["mayusculas"])))
        caracteres_permitidos += caracteres["mayusculas"]

    if minusculas:
        contrase√±a.append(random.choice(list(caracteres["minusculas"])))
        caracteres_permitidos += caracteres["minusculas"]

    if numeros:
        contrase√±a.append(random.choice(list(caracteres["numeros"])))
        caracteres_permitidos += caracteres["numeros"]

    if signos:
        contrase√±a.append(random.choice(list(caracteres["signos"])))
        caracteres_permitidos += caracteres["signos"]

    # Rellenar hasta alcanzar la longitud indicada
    while len(contrase√±a) < longitud:
        contrase√±a.append(random.choice(list(caracteres_permitidos)))

    # Mezclar para que no sea predecible
    random.shuffle(contrase√±a)

    # Convertir la lista en string y copiar al portapapeles
    resultado = "".join(contrase√±a)
    pyperclip.copy(resultado)
    
    return resultado

print(contrase√±a_segura(12))
print("Contrase√±a copiada al portapapeles")
```

# 2. Procesamiento de datos simple

Un compa√±ero de trabajo nos ha pedido un favor, aprovechando el hecho de que ella sabe que estamos aprendiendo a programar. Tiene una historia de partidos de f√∫tbol catal√°n en un archivo, donde se almacenan los nombres de los equipos y los resultados. Necesita procesar los datos autom√°ticamente, para extraer los resultados que necesita.

Utilice el archivo "historic_partites.txt"

Necesitas un programa que devuelva:

    El n√∫mero total de goles que cada equipo ha marcado.
    El nombre del equipo m√°s goleador.
    El nombre del equipo m√°s goleador
    La clasificaci√≥n general (cada victoria: 3 pts, 1 pts empate, derrota 0 pts)


```python
import pandas as pd 

# Leer el archivo
historico = pd.read_csv(r"C:\Users\GCCM_\OneDrive\Desktop\Bootcamp\historic partits.txt",sep= "\t", header=None)

# Nombrar las columnas
historico.columns = ['EquipoLocal', 'Marcador', 'EquipoVisitante']

# Separar goles locales y visitantes 
historico[['GolesLocal', 'GolesVisitante']] = historico['Marcador'].str.split('-', expand=True)

# Convertir los goles a enteros
historico['GolesLocal'] = historico['GolesLocal'].astype(int)
historico['GolesVisitante'] = historico['GolesVisitante'].astype(int)
historico
```

```python
# Obtener la lista de los equipos
historico['EquipoLocal'].unique()
```

```python
equipos = ['Manlleu', 'Vilafranca', 'Olot', 'Cornell√†', 'Cerdanyola',
       'Sant Andreu', 'FC Barcelona', 'Granollers', 'Prat', 'Llagostera',
       'Figueres', 'Europa', 'Terrassa', 'Reus Deportiu', 'Girona FC',
       'Badalona', 'Sabadell', 'Lleida Esportiu', 'N√†stic de Tarragona',
       'RCD Espanyol']

# Diccionarios para calculos  
goles_totales = {}         
goles_encajados = {}       
puntos = {}                
ganados = {}               
empatados = {}             
perdidos = {}             

for equipo in equipos:
    goles_totales[equipo] = 0
    goles_encajados[equipo] = 0
    puntos[equipo] = 0
    ganados[equipo] = 0
    empatados[equipo] = 0
    perdidos[equipo] = 0

# Recorrer cada partido y actualizar datos
for index, partido in historico.iterrows():
    local = partido['EquipoLocal']
    visitante = partido['EquipoVisitante']
    goles_local = partido['GolesLocal']
    goles_visitante = partido['GolesVisitante']

    # Sumar goles a favor y en contra
    goles_totales[local] += goles_local
    goles_totales[visitante] += goles_visitante
    goles_encajados[local] += goles_visitante
    goles_encajados[visitante] += goles_local

    # Asignar puntos y contar partidos ganados/empatados/perdidos
    if goles_local > goles_visitante:
        puntos[local] += 3
        ganados[local] += 1
        perdidos[visitante] += 1
    elif goles_local < goles_visitante:
        puntos[visitante] += 3
        ganados[visitante] += 1
        perdidos[local] += 1
    else:
        puntos[local] += 1
        puntos[visitante] += 1
        empatados[local] += 1
        empatados[visitante] += 1

# Crear un DataFrame final de clasificaci√≥n
clasificacion = pd.DataFrame({
    'partidos ganados': pd.Series(ganados),
    'partidos empatados': pd.Series(empatados),
    'partits perdidos': pd.Series(perdidos),
    'goles': pd.Series(goles_totales),
    'goles encajados': pd.Series(goles_encajados),
    'puntos': pd.Series(puntos)
})

# Ordenar puntos de mayor a menor 
clasificacion = clasificacion.sort_values(by='puntos', ascending=False)

clasificacion
```
