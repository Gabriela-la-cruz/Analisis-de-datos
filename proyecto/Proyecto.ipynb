# Evolución de la natalidad y su relación con el desequilibrio en los nacimientos por sexo

Este análisis examina cómo ha evolucionado la relación entre nacimientos de niños y niñas en una muestra de 50 países, con especial atención al caso de China. Además, se analiza la evolución de la natalidad a lo largo del tiempo para identificar posibles cambios demográficos en las últimas décadas.

El indicador utilizado para analizar el desequilibrio en los nacimientos representa el número de niños nacidos por cada 100 niñas. En condiciones naturales, este valor suele situarse alrededor de 105, por lo que desviaciones respecto a este nivel pueden indicar la existencia de desequilibrios en los nacimientos.

El objetivo del análisis es identificar patrones demográficos, comparar diferentes países y analizar estas desviaciones en el contexto de la reducción de la natalidad observada en muchos países durante las últimas décadas.

## Limpieza y preparación del dataset

```python
# Eliminar columnas vacías tipo Unnamed
df_sex = df_sex.loc[:, ~df_sex.columns.str.contains('Unnamed')]
```

```python
# Detectar columnas de años
years = [col for col in df_sex.columns if col.isdigit()]
```

```python
# Seleccionar solo las columnas necesarias
df_sex = df_sex[["Country Name","Country Code"] + years]
df_sex.head()
```

```python
df_fertility = df_fertility[["Country Name","Country Code"] + years]
df_fertility.head()
```

```python
# Convertir a formato largo

df_sex_long = df_sex.melt(
    id_vars=["Country Name","Country Code"],
    value_vars=years,
    var_name="year",
    value_name="sex_ratio"
)

df_sex_long["year"] = df_sex_long["year"].astype(int)
df_sex_long.head()
```

```python
df_fertility_long = df_fertility.melt(
    id_vars=["Country Name","Country Code"],
    value_vars=years,
    var_name="year",
    value_name="fertility_rate"
)

df_fertility_long["year"] = df_fertility_long["year"].astype(int)
df_fertility_long.head()
```

```python
# Crear variables analíticas
df_sex_long['sex_ratio_per_100_female'] = df_sex_long['sex_ratio'] * 100
df_sex_long['deviation_from_105'] = df_sex_long['sex_ratio_per_100_female'] - 105
```

```python
# Cargar el metadata
ruta_meta = r"C:\Users\GCCM_\OneDrive\Desktop\Bootcamp\proyecto\API_SP.POP.BRTH.MF_DS2_en_csv_v2_8393\Metadata_Country_API_SP.POP.BRTH.MF_DS2_en_csv_v2_8393.csv"
meta = pd.read_csv(ruta_meta)
meta.head()
```

```python
# Filtrar solo países reales
meta_paises = meta[meta['Region'].notna() & (meta['Region'] != 'Aggregates')].copy()

iso_paises = set(meta_paises['Country Code'])
```

```python
# Filtrar países en ambos datasets
df_sex_paises = df_sex_long[df_sex_long["Country Code"].isin(iso_paises)].copy()

df_fertility_paises = df_fertility_long[df_fertility_long["Country Code"].isin(iso_paises)].copy()
```

```python
# Seleccion de paises
my_50 = [
    "CHN","IND","KOR","VNM","NPL","PAK","BGD","AZE","ARM",
    "ALB","GEO","MNE","SRB","BIH","UZB",
    "USA","IDN","BRA","NGA","MEX","RUS","JPN","ETH","PHL","EGY",
    "TUR","IRN","DEU","THA","FRA","GBR","ITA","ZAF","COL","ARG",
    "KEN","GHA","MAR","TZA","SEN",
    "CHL","PER","CAN","VEN","DOM",
    "ESP","SWE","POL",
    "AUS","NZL"
]

df_sex_50 = df_sex_paises[df_sex_paises['Country Code'].isin(my_50)].copy()


df_fertility_50 = df_fertility_paises[df_fertility_paises["Country Code"].isin(my_50)].copy()
```

```python
# Eliminar años sin datos reales
df_sex_50 = df_sex_50[df_sex_50["year"] <= 2023]
```

```python
df_fertility_50 = df_fertility_50[df_fertility_50["year"] <= 2023]
```

```python
# Unir datasets
df_combined = pd.merge(
    df_sex_50,
    df_fertility_50,
    on=["Country Name","Country Code","year"],
    how="left"
)

df_combined.head()
```

```python
# Crear variable específica para periodos de la política china
def china_policy_phase(country, year):
    if country != 'CHN':
        return 'not_applicable'
    elif year <= 1978:
        return 'pre_policy'
    elif year <= 2015:
        return 'one_child_policy'
    else:
        return 'post_policy'
    
df_combined["china_policy_phase"] = df_combined.apply(
    lambda row: china_policy_phase(row["Country Code"], row["year"]),
    axis=1
)
```

## Validación del dataset

```python
# Validación final del dataset combinado

# Ordenar dataset
df_combined = df_combined.sort_values(['Country Code', 'year']).reset_index(drop=True)

# Comprobar número de países y dimensiones
df_combined['Country Code'].nunique(), df_combined.shape

# Verificar valores faltantes
df_combined.isna().sum()

# Ver columnas finales
df_combined.columns

df_combined.head()
```

## Indicadores utilizados

En este análisis se emplean los siguientes indicadores para estudiar la natalidad y el equilibrio en los nacimientos:

- **Número de niños nacidos por cada 100 niñas (sex ratio)**:  
  Indicador principal que permite identificar posibles desequilibrios en los nacimientos. En condiciones naturales, este valor se sitúa en torno a 105.

- **Tasa de fertilidad (fertility rate)**:  
  Representa el número promedio de hijos por mujer y permite analizar la evolución de la natalidad a lo largo del tiempo.

- **Desviación respecto al valor esperado (~105)**:  
  Permite medir en qué medida el sex ratio se aleja del nivel biológico esperado.

Estos indicadores permiten analizar tanto el equilibrio en los nacimientos como la evolución de la natalidad, así como explorar posibles relaciones entre ambos fenómenos.

## Análisis exploratorio de datos (EDA)

En esta sección se realiza un análisis exploratorio de los datos con el objetivo de estudiar la evolución de la natalidad y el equilibrio en los nacimientos, así como identificar posibles relaciones entre ambos fenómenos.

Para ello, se analizan tendencias globales, diferencias entre países y casos específicos que permiten comprender mejor el comportamiento de los indicadores.

#### 1. Análisis global de la fertilidad y el desequilibrio en los nacimientos

```python
# Resumen estadístico general
df_combined.describe().round(2)
```

```python
# Evolución global conjunta
df_combined.groupby("year")[["fertility_rate","sex_ratio_per_100_female"]] \
.mean().round(2)
```

```python
country_corr = (
    df_combined.groupby("Country Name")[["fertility_rate","sex_ratio_per_100_female"]]
    .corr()
    .unstack()
)

country_corr.head()
```

Los resultados muestran que la tasa de fertilidad presenta una clara tendencia descendente a nivel global, pasando de valores superiores a 5 hijos por mujer en los años sesenta a valores cercanos a 2 en la actualidad.

Por otro lado, el número de niños nacidos por cada 100 niñas se mantiene relativamente estable en torno al valor biológico esperado (~105), con variaciones moderadas a lo largo del tiempo.

La correlación entre ambas variables es débil, lo que indica que la disminución de la natalidad no se asocia de forma directa con un aumento del desequilibrio en los nacimientos a nivel global.

#### 2. Comparación entre países

```python
# Medias por país 
country_means = (
    df_combined.groupby("Country Name")[["fertility_rate","sex_ratio_per_100_female"]]
    .mean()
    .round(2)
)

# Países con mayor desequilibrio
country_means.sort_values("sex_ratio_per_100_female", ascending=False).head(10)
```

```python
# Países más cercanos al equilibrio
country_means.sort_values("sex_ratio_per_100_female").head(10)
```

Los resultados muestran diferencias claras entre países en el número de niños nacidos por cada 100 niñas.

Los valores más elevados se concentran principalmente en países de Asia y Europa del Este, como China, Azerbaiyán o Armenia, donde se observan desequilibrios significativos respecto al valor biológico esperado (~105).

Por el contrario, los países más cercanos al equilibrio se encuentran en su mayoría en África y América Latina, como Kenya, Tanzania o Senegal, donde el indicador se mantiene próximo al nivel natural.

Estos resultados indican que el desequilibrio en los nacimientos no es un fenómeno global, sino que se concentra en contextos específicos, lo que sugiere la influencia de factores sociales, culturales o institucionales.

#### 3. Análisis de países representativos

```python
# Selección de países representativos
selected_countries = ["CHN", "KOR", "ESP", "USA", "KEN", "VEN"]

df_selected = df_combined[df_combined["Country Code"].isin(selected_countries)].copy()

# Resumen inicio vs final
summary_selected = (
    df_selected.sort_values("year")
    .groupby("Country Name")
    .agg(
        fertilidad_inicio=("fertility_rate", "first"),
        fertilidad_final=("fertility_rate", "last"),
        sexratio_inicio=("sex_ratio_per_100_female", "first"),
        sexratio_final=("sex_ratio_per_100_female", "last")
    )
    .round(2)
)

# Cálculo de cambios
summary_selected["cambio_fertilidad"] = (
    summary_selected["fertilidad_final"] - summary_selected["fertilidad_inicio"]
)

summary_selected["cambio_sexratio"] = (
    summary_selected["sexratio_final"] - summary_selected["sexratio_inicio"]
)

# Redondear resultados finales
summary_selected = summary_selected.round(2)

summary_selected
```

El análisis de países representativos permite identificar patrones diferenciados en la relación entre fertilidad y desequilibrio en los nacimientos.

China muestra una fuerte caída de la fertilidad acompañada de un aumento significativo del sex ratio, lo que indica un desequilibrio creciente en los nacimientos. En contraste, Corea del Sur presenta una reducción de la fertilidad junto con una disminución del desequilibrio, lo que sugiere una corrección progresiva del fenómeno.

En otros países como España o Estados Unidos, la fertilidad también disminuye, pero el sex ratio se mantiene estable en valores cercanos al nivel natural. Por su parte, Kenya presenta una caída importante de la fertilidad sin apenas cambios en el equilibrio de los nacimientos, manteniéndose dentro del rango biológico esperado. De hecho, valores alrededor de 105–106 se consideran normales a nivel internacional 

Estos resultados refuerzan la idea de que la disminución de la natalidad no implica necesariamente un aumento del desequilibrio, sino que depende de factores específicos de cada país.

```python
# ¿Existe una tendencia global en el desequilibrio de los nacimientos por sexo?

# Promedio global
global_sex = df_combined.groupby("year")["sex_ratio_per_100_female"].mean().reset_index()
global_sex["Country Name"] = "Global"

# Países seleccionados
countries = ["CHN", "KOR", "USA", "VEN", "ESP", "IND","AZE"]

country_data = (
    df_combined[df_combined["Country Code"].isin(countries)]
    [["year", "Country Name", "sex_ratio_per_100_female"]]
    .copy()
)

# Unir datos globales y países
plot_data_global = pd.concat([
    global_sex.rename(columns={"sex_ratio_per_100_female": "value"})[["year", "Country Name", "value"]],
    country_data.rename(columns={"sex_ratio_per_100_female": "value"})
])

# Gráfico
fig = px.line(
    plot_data_global,
    x="year",
    y="value",
    color="Country Name",
    title="Evolución del número de niños nacidos por cada 100 niñas (Global y países seleccionados)",
    labels={
        "year": "Año",
        "value": "Niños por cada 100 niñas",
        "Country Name": "País"
    }
)

fig.add_hline(
    y=105,
    line_dash="dot",
    line_color="gray",
    opacity=0.6,
    annotation_text="Valor esperado (~105)",
    annotation_position="top right"
)

fig.write_html("grafico_1_sexratio_global_paises.html")
fig.show()
```

```python
# ¿Está cayendo la natalidad a nivel mundial? 

global_fert = df_combined.groupby("year")["fertility_rate"].mean().reset_index() 

fig = px.line( 
    global_fert, 
    x="year", 
    y="fertility_rate", 
    title="Evolución global de la tasa de fertilidad", 
    labels={ "year": "Año", "fertility_rate": "Hijos por mujer" 
        } 
    ) 

fig.write_html("grafico_2_fertilidad_global.html") 
fig.show()
```

```python
# ¿Qué países presentan desequilibrios en los nacimientos y cuáles se mantienen cerca del nivel biológico esperado?

# Lista de países
countries = ["CHN", "KOR", "USA", "VEN", "ESP", "IND", "AZE", "KEN"]

# Filtrar + agrupar
plot_data_points = (
    df_combined[df_combined["Country Code"].isin(countries)]
    .groupby(["Country Name", "Country Code"], as_index=False)["sex_ratio_per_100_female"]
    .mean()
)

# Calcular desviación respecto a 105
plot_data_points["desviacion"] = abs(
    plot_data_points["sex_ratio_per_100_female"] - 105
)

# Tamaño con base mínima visible
plot_data_points["size"] = plot_data_points["desviacion"] * 2 + 6

fig = px.scatter(
    plot_data_points,
    x="sex_ratio_per_100_female",
    y="Country Name",
    size="size",
    size_max=24,
    hover_data={
        "size": False 
    },
    title="Comparación del número de niños nacidos por cada 100 niñas por país",
    labels={
        "sex_ratio_per_100_female": "Niños por cada 100 niñas",
        "Country Name": "País"
    }
)

fig.add_vline(
    x=105,
    line_dash="dash",
    annotation_text="Valor esperado (~105)"
)

fig.update_traces(
    marker=dict(
        line=dict(width=1.5, color="darkblue"),
        opacity=0.85
    )
)

fig.update_layout(
    margin=dict(l=160, r=40, t=60, b=60)
)

fig.write_html("grafico_3_puntos_desequilibrio.html")
fig.show()
```

```python
# ¿Qué países han aumentado más su sex ratio con el tiempo?
change = (
    df_combined.sort_values("year")
    .groupby("Country Name")
    .agg(
        inicio=("sex_ratio_per_100_female", "first"),
        fin=("sex_ratio_per_100_female", "last")
    )
)

change["cambio"] = change["fin"] - change["inicio"]

# Top países con mayor aumento
top_change = (
    change.sort_values("cambio", ascending=False)
    .head(10)
    .reset_index()
)

import plotly.express as px

fig = px.bar(
    top_change,
    x="cambio",
    y="Country Name",
    orientation="h",
    title="Países con mayor aumento del número de niños por cada 100 niñas (1960–2023)",
    labels={
        "cambio": "Cambio en niños por cada 100 niñas",
        "Country Name": "País"
    }
)

fig.write_html("grafico_4_extra_cambio_temporal.html")
fig.show()
```

```python
# ¿Coincide la caída de fertilidad con el aumento del desequilibrio?

countries = ["CHN", "KOR", "ESP", "USA", "VEN","AZE","IND"]

country_names = {
    "CHN": "China",
    "KOR": "Corea del Sur",
    "ESP": "España",
    "USA": "Estados Unidos",
    "VEN": "Venezuela",
    "AZE": "Azerbaijan",
    "IND": "India"
}

# crear figura
fig = go.Figure()

# añadir todas las trazas pero ocultas
for code in countries:
    data = (
        df_combined[df_combined["Country Code"] == code]
        .groupby("year")[["fertility_rate", "sex_ratio_per_100_female"]]
        .mean()
        .reset_index()
    )

    # fertilidad
    fig.add_trace(go.Scatter(
        x=data["year"],
        y=data["fertility_rate"],
        mode="lines",
        name=f"{country_names[code]} - Fertilidad",
        visible=(code == "CHN")  # solo China visible al inicio
    ))

    # sex ratio
    fig.add_trace(go.Scatter(
        x=data["year"],
        y=data["sex_ratio_per_100_female"],
        mode="lines",
        name=f"{country_names[code]} - Sex ratio",
        yaxis="y2",
        visible=(code == "CHN")
    ))

# crear botones dropdown
buttons = []

for i, code in enumerate(countries):
    visibility = [False] * (len(countries) * 2)
    
    # activar solo las dos trazas del país seleccionado
    visibility[i*2] = True
    visibility[i*2 + 1] = True

    buttons.append(dict(
        label=country_names[code],
        method="update",
        args=[{"visible": visibility}]
    ))

fig.update_layout(
    title="Relación entre fertilidad y desequilibrio por país",
    xaxis_title="Año",
    yaxis=dict(title="Hijos por mujer"),
    yaxis2=dict(
        title="Niños por cada 100 niñas",
        overlaying="y",
        side="right",
        range=[101, 119]
    ),
    updatemenus=[
        dict(
            buttons=buttons,
            direction="down",
            showactive=True,
            x=0.1,
            y=1.15
        )
    ]
)

fig.write_html("grafico_8_dropdown_paises.html")
fig.show()
```

## Visualización interactiva (Flourish)

Con el objetivo de representar la evolución temporal de los datos de forma dinámica, se generan datasets para visualizaciones en Flourish.

```python
# 1. Evolución del desequilibrio en los nacimientos por país

# Países seleccionados
countries = [
    "China", "Korea, Rep.", "India",
    "United States", "Spain", "Kenya", "Venezuela, RB","Azerbaijan"
]

# Filtrar
df_flourish = df_combined[df_combined["Country Name"].isin(countries)].copy()

# Cambiar nombres conflictivos
df_flourish["Country Name"] = df_flourish["Country Name"].replace({
    "Korea, Rep.": "South Korea",
    "Venezuela, RB": "Venezuela"
})

# Pivot a formato ancho
flourish_race = df_flourish.pivot(
    index="Country Name",
    columns="year",
    values="sex_ratio_per_100_female"
).reset_index()

# Renombrar columna principal
flourish_race = flourish_race.rename(columns={"Country Name": "Country"})

# Redondear para que quede limpio
year_cols = [col for col in flourish_race.columns if col != "Country"]
flourish_race[year_cols] = flourish_race[year_cols].round(1)

# Exportar a Excel
flourish_race.to_excel("flourish_sexratio_race.xlsx", index=False)

flourish_race.head()
```
